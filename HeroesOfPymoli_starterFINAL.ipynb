```python
#Retrieve the item ID, Item Name and Item Price Colums
item_temp = pdc.loc[:,["Item ID", "Item Name", "Price"]]
```

```python
#Group by Item ID
item_g = item_temp.groupby(["Item ID", "Item Name"])

#Calculate average price per item
avg_price = item_g['Price'].mean()
```

```python
#Calculate the number of count per item
other_count = item_g['Price'].count()
```

```python
#Total Purchase Value
total_purchase = item_g["Price"].sum()
```
